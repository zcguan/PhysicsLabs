```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
c = 10**(-6)
r1=r2=10**3
v0=1
tau=(r1+r2)*c*100
```

```python
I = lambda t: v0/(r1+r2)*np.exp(-t/tau)*10**3
Q = lambda t: c*v0*(1-np.exp(-t/tau))*10**6
```

```python
x = np.linspace(0,tau*5)
```

```python
plt.plot(x,I(x),label="Current")
plt.plot([tau]*len(x),np.linspace(0,I(tau)),label=r"t = $\tau$")
plt.plot(tau,I(tau),'o')
plt.annotate(r'$\tau=$'+ str(tau),(tau,I(tau)), textcoords="offset points", xytext=(10,0),ha='left')
plt.xlabel("Time(cs)")
plt.ylabel("I (mA)")
plt.title("Current vs time")
plt.legend()
plt.savefig("I")
plt.show()
```

```python
plt.plot(x,Q(x),label="Charge")
plt.plot([tau]*len(x),np.linspace(0,Q(tau)),label=r"t = $\tau$")
plt.plot(tau,Q(tau),'o')
plt.annotate(r'$\tau=$'+ str(tau),(tau,Q(tau)), textcoords="offset points", xytext=(10,0),ha='left')
plt.xlabel("Time(cs)")
plt.ylabel("Q ($\mu$C)")
plt.title("Charge vs time")
plt.legend()
plt.savefig("Q")
plt.show()
```
