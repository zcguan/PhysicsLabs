```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
v0=1
T=10**-3
L=1
R=1000
tau = L/R
tau
```

```python
I = lambda t: (v0/(L*T))*((tau*t-tau**2)+tau**2*np.exp(-t/tau))
It = lambda t: (v0/(L*T))*((tau*t-tau**2))
```

```python
x = np.linspace(0,10**-2/1.5,100)
y = I(x)
yt = It(x)

for i in range(len(x)):
    if abs(y[i]-yt[i])<0.0001: 
        print(i)
        break
x[i]
```

```python
plt.plot(x,y)
plt.xlabel('Time ')
plt.ylabel('Current')
plt.title('RL circuit driven by linear voltage')
plt.savefig('1.jpg')
```

```python
A=lambda x: (1/(np.sqrt(1+x**-2)))
phi=lambda x: np.arctan(x**-1)
```

```python
x = (np.linspace(10**-3,10**3,10**5))
ya = 20*np.log10(A(x))
yp = (phi(x))
```

```python
plt.plot(x,ya)
plt.plot([1]*len(x),np.linspace(min(ya),max(ya),len(x)), '--')
plt.xscale('log')
plt.xlabel('$\omega\\tau$ (log scale)')
plt.ylabel('Amplitude of $V_0\'/V_0$ in dB')
plt.title('RL driven by sinusoidal voltage')
plt.savefig('2.jpg')
```

```python
plt.plot(x,(yp))
plt.plot([1]*len(x),np.linspace(min(yp),max(yp),len(x)), '--')
plt.xscale('log')
plt.xlabel('$\omega\\tau$ (log scale)')
plt.ylabel(r'$\varphi$')
plt.title('Phase offset of V\'')
plt.savefig('3.jpg')
```
