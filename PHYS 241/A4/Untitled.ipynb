```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = 100
r = [0.1, 1, 10]
C = 10**-5
L = 10**-3
V = 1
w = 1/np.sqrt(L*C)
```

```python
f = lambda x,i: R*np.sqrt(((1-x**2)**2+(w*x*r[i]*C)**2)/((R*(1-x**2)+r[i])**2+(w*x)**2*(r[i]*R*C+L)**2))
```

```python
x = np.linspace(0,2,10**4)
plt.plot(x,f(x,0), label = 'r = 0.1 $\Omega$')
plt.plot(x,f(x,1), label = 'r = 1 $\Omega$')
plt.plot(x,f(x,2), label = 'r = 10 $\Omega$')
plt.title('Normalized current amplitude',fontsize = 15)
plt.xlabel('$\omega/\omega_0$',fontsize = 15)
plt.ylabel('$I(t)/I_{max}$',fontsize = 15)
plt.legend()
plt.savefig('q4.jpg')
```
