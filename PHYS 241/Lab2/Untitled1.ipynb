```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
tri = lambda x,p2pA,L,d: abs((2*p2pA/L)*(x%L)-p2pA)-p2pA/2+d
sqr = lambda x,p2pA,L,d: (p2pA if (x%L) < L/2 else 0)-p2pA/2+d
```

```python
x = np.linspace(-50,50,10**3)
ytri = -tri(x,15,20,0)
plt.axhline(linewidth=1,ls='--',color='k')
plt.axvline(linewidth=1,ls='--',color='k')
plt.plot(x,ytri)
plt.ylim(-8,8)
plt.grid()
plt.xlabel('Time in $\mu$s')
plt.ylabel('Voltage in mV')
plt.title('$V_C$ voltage')
plt.savefig('q1vc.jpg')
```

```python
x = np.linspace(-40,40,10**4)
ysqr = [sqr(n,2,20,0) for n in x]
plt.plot(x,ysqr)
plt.axhline(linewidth=1,ls='--',color='k')
plt.axvline(linewidth=1,ls='--',color='k')
plt.grid()
plt.ylim(-1.5,1.5)
plt.xlabel('Time in $\mu$s')
plt.ylabel('Voltage in V')
plt.title('Generator voltage')
plt.savefig('q1gen.jpg')
```

```python
x = np.linspace(-200,200,10**4)
ytri = tri((x-25),2,100,0)
plt.plot(x,ytri)
plt.grid()
plt.axhline(linewidth=1,ls='--',color='k')
plt.axvline(linewidth=1,ls='--',color='k')
plt.xlabel('Time in ms')
plt.ylabel('Voltage in V')
plt.title('Input voltage')
plt.savefig('q3in.jpg')
```
