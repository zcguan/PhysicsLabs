```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
vi = 5
fh = 1600
```

```python
e2 = np.loadtxt('e2datat2.csv', delimiter=',').transpose()
e4 = np.loadtxt('e4data.csv', delimiter=',').transpose()
```

```python
x = list(map(lambda f: np.log10(f/fh), e2[0]))
y = list(map(lambda v: 20*np.log10(v/vi), e2[1]))
xf = x[:7]
yf = y[:7]
m,a = np.polyfit(xf,yf,1)
xf = np.linspace(min(xf),.1)

#plt.grid()
plt.plot(x,y,'o')
plt.axvline(linewidth=1,color = 'k',ls='--')
plt.axhline(linewidth=1,y=-3,color='k',ls='--',label='-3 db')
plt.plot(xf,list(map(lambda x: m*x+a, xf)))
plt.plot(x[8:],[0]*len(x[8:]))
plt.title('Bode plot')
plt.xlabel('log f/f$_H$')
plt.ylabel('H$_{db}$')
plt.legend()
plt.savefig('e2.jpg')
```

```python
x = list(map(lambda f: np.log10(f/40500), e4[0]))
y = list(map(lambda v: 20*np.log10(v/vi), e4[1]))
xf = x[15:]
yf = y[15:]
m,a = np.polyfit(xf,yf,1)
xf = np.linspace(-.1,max(xf))

#plt.grid()
plt.plot(x,y,'o')
plt.plot(xf,list(map(lambda x: m*x+a, xf)))
plt.plot(x[:15],[0]*15)
plt.axvline(linewidth=1,color = 'k',ls='--')
plt.axhline(linewidth=1,y=-3,color='k',ls='--', label='-3 db')
plt.title('Bode plot')
plt.xlabel('log f/f$_H$')
plt.ylabel('H$_{db}$')
plt.legend()
plt.savefig('e4.jpg')
```
