```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
exp1f,exp1g, exp1p = np.loadtxt('Experiment1.csv', delimiter=',')
exp2f,exp2g, exp2p  = np.loadtxt('Experiment2NoSpoiler.csv', delimiter=',')
exp2fs,exp2gs, exp2ps  = np.loadtxt('Experiment2Spoiler.csv', delimiter=',')
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.plot(exp1f,exp1p,'.')
ax2.plot(exp1f,exp1g,'.')
```

```python
imax = max(exp1g)
y = exp1g.tolist()
f0 = exp1f[y.index(imax)+1]
y3 = imax/np.sqrt(2)
x3 = []
for i,n in enumerate(y):
    if abs(n-y3) < 0.1:
        x3.append(exp1f[i])

print(y3,x3)
width = x3[1]-x3[0]
# for x in x3:
#     print((2*np.pi*x*50*10**(-3)) - (1/(2*np.pi*x*0.005*10**(-6))))

print('f0',f0)
print(width)
q = f0/width
r = (2*np.pi *f0 * 48.9*10**(-3))/q
print('r=',r)
print('Q=',q)
        
plt.plot(exp1f,exp1g,'.')
plt.plot(f0,imax,'ko',label = '$peak$')
plt.plot(x3,[y3,y3],'o',label = '$1/\sqrt{2}$ amplitude')
plt.xlabel('frequency in Hz')
plt.ylabel('Gain $V_C/V_i$')
plt.title('Resonance curve')
plt.legend()
plt.savefig('exp1.png')
```
